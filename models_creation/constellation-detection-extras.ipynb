# Preprocessing used kernel visualizer

```python
import matplotlib.pyplot as plt
import numpy as np

kernel = np.array([[0.00390625, 0.015625, 0.0234375, 0.015625, 0.00390625],
                   [0.015625, 0.0625, 0.09375, 0.0625, 0.015625],
                   [0.0234375, 0.09375, 0.140625, 0.09375, 0.0234375],
                   [0.015625, 0.0625, 0.09375, 0.0625, 0.015625],
                   [0.00390625, 0.015625, 0.0234375, 0.015625, 0.00390625]])

fig, ax = plt.subplots()
heatmap = ax.pcolor(kernel, cmap='viridis')

for i in range(len(kernel)):
    for j in range(len(kernel[i])):
        ax.text(j + 0.5, i + 0.5, f'{kernel[i][j]:.3f}',
                ha='center', va='center', color='black')

cbar = plt.colorbar(heatmap)
ax.set_title('5x5 Gaussian Kernel')
plt.show()
```
